```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import missingno as msno

df = pd.read_csv('../../Dataset/PCA_Dataset_v1.csv', encoding='latin-1')
df = df.replace(r'\s+', np.nan, regex=True)
df = df.replace(r'?', np.nan)
df = df.replace("nan",np.nan)
df = df.dropna(subset=['History of kidney disease'])

msno.matrix(df)
plt.savefig('images/entireMissingData.png')
```

```python
df_temp = df.iloc[:,50:100]
msno.matrix(df_temp)
plt.savefig('images/entireMissingData2.png')
```

```python
df_temp = df.iloc[:,60:100]
msno.bar(df_temp)
plt.savefig('images/entireNullity.png')
```

```python
msno.heatmap(df)
plt.savefig('images/entireHeatmap.png')
```

```python
msno.dendrogram(df)
plt.savefig('images/entireDendrogram.png')
```

```python
df = pd.read_csv('../../Dataset/Kidney_v2.csv', encoding='latin-1')
df = df.replace(r'\s+', np.nan, regex=True)
df = df.replace(r'?', np.nan)
df = df.replace("nan",np.nan)
df = df.dropna(subset=['History of kidney disease'])

df_temp1 = df.iloc[:,0:50]
df_temp2 = df.iloc[:,50:100]
df_temp3 = df.iloc[:,100:116]

msno.matrix(df_temp1)
plt.savefig('images/kidneyMissingData1.png')
msno.matrix(df_temp2)
plt.savefig('images/kidneyMissingData2.png')
msno.matrix(df_temp3)
plt.savefig('images/kidneyMissingData3.png')
```

```python
msno.heatmap(df_temp1)
plt.savefig('images/kidneyHeatmap.png')
```

```python
msno.dendrogram(df)
plt.savefig('images/kidneyDendrogram.png')
```
